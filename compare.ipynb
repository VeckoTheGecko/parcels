```python
# Use xarray to open `data-v4/nemo/field/*U*.nc`

from pathlib import Path

import xarray as xr

from parcels._datasets import utils

PATH = Path().parent / "data-v4"


def open_nemo() -> xr.Dataset:
    # xr openmfdataset concat dim is time_counter and combine is nested

    return xr.open_mfdataset(
        f"{PATH}/nemo/field/*U*.nc",
        concat_dim="time_counter",
        combine="nested",
    )


reference = open_nemo()
```

```python
print(reference)
print(reference.attrs)
```

```python
print("\nDataset Attributes:")
print(reference.attrs)

print("\nCoordinates Attributes:")
for coord_name in reference.coords:
    print(f"  {coord_name}: {reference[coord_name].attrs}")

print("\nData Variables Attributes:")
for var_name in reference.data_vars:
    print(f"  {var_name}: {reference[var_name].attrs}")
```
